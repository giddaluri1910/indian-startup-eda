```python
import pandas as pd
df = pd.read_csv('startup_funding_sample.csv')
df.head()
```

```python
# Check the structure of the dataset
df.info()
```

```python
# See summary statistics for numeric columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop rows with missing values (if any)
df.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract year
df['Year'] = df['Date'].dt.year

# Total funding per year
funding_by_year = df.groupby('Year')['AmountInUSD'].sum()

# Plot
funding_by_year.plot(kind='bar', color='skyblue')
plt.title("Total Funding by Year")
plt.xlabel("Year")
plt.ylabel("Amount (USD)")
plt.grid(True)
plt.show()
```

```python
top_cities = df['CityLocation'].value_counts()

top_cities.plot(kind='bar', color='orange')
plt.title("Number of Startups by City")
plt.xlabel("City")
plt.ylabel("Number of Startups")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
